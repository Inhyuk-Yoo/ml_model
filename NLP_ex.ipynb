```python
import pandas as pd
import urllib.request
import nltk
import string
from nltk.corpus import stopwords
nltk.download('stopwords')

train_data = pd.read_csv('./train.csv', encoding = 'utf-8')
test_data = pd.read_csv('./test.csv', encoding = 'utf-8')

train_data.drop_duplicates(subset=['mail'], inplace=True)

label_data = train_data['label']
mail_data = train_data['mail']
mail_test = test_data['mail']
print("type is :")
print(type(mail_data))
print("mail_data data before : ")
print(mail_data.head(2))
print("mail_test data before : ")
print(mail_test.head(2))

def prepro_mail(mail):
    punc = [char for char in mail if char not in string.punctuation]
    punc = ''.join(punc)
    clear = [word for word in punc.split() if word.lower() not in stopwords.words('english')]
    return clear

mail_data = prepro_mail(mail_data)
mail_test = prepro_mail(mail_test)
print("after function pre type is :")
print(type(mail_data))
mail_data = pd.Series(mail_data)
mail_test = pd.Series(mail_test)
print("after series type is :")
print(type(mail_data))
# # list after preprocessing

mail_data = mail_data.head(2)
mail_test = mail_test.head(2)

print("mail_data data after : ")
print(mail_data)
print("mail_test data after : ")
print(mail_test)
```
